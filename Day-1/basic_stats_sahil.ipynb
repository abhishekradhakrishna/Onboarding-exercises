```python
#Basic terms	
#1	Mean, median, mode, range, standard deviation, variance
#2	Population vs sample (mean, variance and std deviation) - Explain difference and reason to choose either of them
#3	Chebyshevs theorem
#4	Central limit theorem

#1	Mean, median, mode, range, standard deviation, variance
import sklearn
import pandas as pd
from sklearn import datasets
import statistics
import numpy as np
import matplotlib.pyplot as plt

iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')

#returns the avg of the list given
mean = statistics.mean(iris.sepal_length)
#plotting it on a graph
y = np.array(iris.sepal_length)
x = np.array([i for i in range(1,len(iris.sepal_length)+1)])
y_mean = [mean]*len(x)
data_line = plt.plot(x,y, label='Data')
mean_line = plt.plot(x,y_mean, label='Mean', linestyle='--')
print('Mean of the data:')
plt.show()

#returns the centremost measure of the list
median = statistics.median(iris.sepal_length)

#returns the most frequently occuring value of the list
mode = statistics.mode(iris.sepal_length)

#sample standard deviation
stdev = statistics.stdev(iris.sepal_length)
#population standard deviation
pstdev = statistics.pstdev(iris.sepal_length)

#sample variance
variance = statistics.variance(iris.sepal_length)
#population variance
pvariance = statistics.pvariance(iris.sepal_length)



#Population standard deviation is the exact parameter value used to measure 
#the dispersion from the center whereas the sample standard deviation is an unbiased estimator for it.
#
#Population standard deviation is calculated when all the data regarding each individual of the population is known.
#In sample standard deviation, division by (n-1) takes place whereas in population, division by (n) takes place.



#3	Chebyshevs theorem
#For normal distributions, about 68% of results will fall between +1 and -1 standard deviations from the mean, 
#about 95% will fall between +2 and -2 standard deviations, the Theorem allows you to use this idea for any distribution, 
#even if that distribution isn’t normal. 
#The theorem states:
#For a population or sample, the proportion of observations is no less than (1 – (1 / k**2 ))


#4	Central limit theorem
#If the population from which the sample has a been drawn is a normal population then the sample means would be equal 
#to population mean and the sampling distribution would be normal. According to Central Limit Theorem, for sufficiently
#large samples with size greater than 30, the shape of the sampling distribution will become more and more like a normal 
#distribution, irrespective of the shape of the parent population. This theorem explains the relationship between the population
#distribution and sampling distribution.
```
